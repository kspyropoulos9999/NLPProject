```python
"""CIS 2532 NET 01
Dr. Shamsuddin
Kostas Spyropoulos
HW 7 
Program performs various uses of NLP"""
import requests
import nltk
from nltk.corpus import stopwords
import imageio
from wordcloud import WordCloud
from textblob import TextBlob
import matplotlib.pyplot as plt
from textblob.sentiments import NaiveBayesAnalyzer
from bs4 import BeautifulSoup
from unidecode import unidecode
from textblob import Word
from spellchecker import SpellChecker
nltk.download('wordnet', quiet=True)

# Exercise 12.1 

print('Exercise 12.1')
print()

# the get function gets the URL argument and sends the URL's webpage as a response object
response = requests.get('https://www.python.org')

# reponse.content possesses web page's composition
response.content 

# BeautifulSoup object is made to analyze text of response.content
soup = BeautifulSoup(response.content,'html5lib')

# text of website is returned without web browser structural info
text = soup.get_text(strip=True) 

# stops is equal to the stop words of the english language
stops = stopwords.words('english')

# textblob of'https://www.python.org'
blob = TextBlob(text)

# list to hold list of words that are not stop wrods
NonStopWords = []

# stop words are elminated from blob.words
# for loop appends words that are not stop words into the NonStopWords list
for wordFound in [word for word in blob.words if word not in stops]:
    NonStopWords.append(str(wordFound))
    

# wordcloud is generated from blob.words with no stop words
# .replace("'", '') is added to remove unecessary apostrophes
wordcloud = WordCloud(colormap='prism', background_color='white').generate(str(NonStopWords).replace("'", ''))

# plot axes are removed
plt.axis('off')

# input wordcloud to be shown
plt.imshow(wordcloud)

plt.show()


# Exercise 12.2

print('Exercise 12.2')
print()


print('Tokenized sentences of https://www.python.org')
print()

# print tokenized sentences of blob
print(blob.sentences)
print()

print('Tokenized words of https://www.python.org')
print()

# print tokenized words of blob
print(blob.words)
print()

print('Extracted noun phrases of https://www.python.org')
print()

# print extracted noun phrases of blob
print(blob.noun_phrases)

# Exercise 12.3

print('Exercise 12.3')
print()

# the get function gets the URL argument and sends the URL's webpage as a Response object
response = requests.get('https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html')

# reponse.content possesses web page's composition
response.content 

# BeautifulSoup object is made to analyze text of response.content
soup = BeautifulSoup(response.content,'html5lib')

# text of website is returned without web browser structural info
text = soup.get_text(strip=True)

# blob is equal to textblob of 'https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html'
blob = TextBlob(text)

print()
print('Overall sentiment of https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html')

# print the sentiment of overall blob
print(blob.sentiment)
print()

print('Sentiment of each sentence of https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html')

# print the sentiment of every sentence in blob
for sentence in blob.sentences:
    print(sentence.sentiment)

# Exercise 12.4

print('Exercise 12.4')
print()

# the get function gets the URL argument and sends the URL's webpage as a Response object
response = requests.get('https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html')

# reponse.content possesses web page's composition
response.content

# BeautifulSoup object is made to analyze text of response.content
soup = BeautifulSoup(response.content,'html5lib')

# text of website is returned without web browser structural info
text = soup.get_text(strip=True)

# blob is equal to textblob of 'https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html'
# and will be analuzed with NaiveBayesAnalyzer
blobNaive = TextBlob(text, analyzer=NaiveBayesAnalyzer())

# print overall sentiment of blobNaive
print("Overall Naive Bayes Sentiment of https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html")
print(blobNaive.sentiment)
print()

# print sentiment of every sentence in blobNaive
print("Sentence Naive Bayes Sentiments of https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html")
for sentenceNaive in blobNaive.sentences:
    print(sentenceNaive.sentiment)

# Exercise 12.5

print()
print('Exercise 12.5')
print()

# read the the Gutenber.txt
with open('C:\\Users\\kspyr\\Documents\\Desktop Stuff\\CIS 2532\\HW 7\\Gutenberg.txt', 'r', encoding="utf8") as text1:
    content = text1.read();
    
# blob is equal to the textblob of the Gutenberg.txt
blob = TextBlob(content)

print("Misspelled words along with correction suggestions")
print()
# for loop iterate through all of the tokenized words of the blob and displays possible corrections for each misspelled word
for wordFound in blob.words:
    stringSpellCheck = str(Word(wordFound).spellcheck())
    stringZero = '0.0)'
    
    # if the spellcheck of the word of the blob being evaluated does not possess str(wordFound) and stringZero,
    # the word of the blob being evaluated must print the word of the blob as well as the correction suggestions.
    
    # str(wordFound) must be absent from the word of the blob being evaluated in order to execute the code under the if statement
    # because the presence of str(wordFound) in stringSpellCheck means the correction suggestion is the word itself which means the correction suggestion
    # is not truly a correction suggestion. In order to prevent such errors, str(wordFound) must be absent from the word of the blob being evaluated
    # to then execute the code under the if statement.
    
    # stringTwo must be absent from the word of the blob being evaluated in order to execute the code under the if statement
    # because the presence of stringTwo or '0.0)' in string represents that the correction suggestion's chance of being right is 0 which means
    # the supposed correction suggestion is not really a correction suggestion. In order to prevent such errors, '0.0)' must be absent from the word of the blob being evaluated
    # to then execute the code under the if statement.

    if str(wordFound) not in stringSpellCheck and stringZero not in stringSpellCheck:
        print()
        print('Word in error: ' + str(wordFound))
        print('Correction suggestions:')
        print(Word(wordFound).spellcheck())  
```
