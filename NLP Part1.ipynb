```python
"""CIS 2532 NET 01
Dr. Shamsuddin
Kostas Spyropoulos
HW 7 
Program performs various uses of NLP
Exercises 12.1 & 12.2"""
import requests
import nltk
from nltk.corpus import stopwords
import imageio
from wordcloud import WordCloud
from textblob import TextBlob
import matplotlib.pyplot as plt
from textblob.sentiments import NaiveBayesAnalyzer
from bs4 import BeautifulSoup
from unidecode import unidecode
from textblob import Word
from spellchecker import SpellChecker

# Exercise 12.1 

print('Exercise 12.1')
print()

# the get function gets the URL argument and sends the URL's webpage as a response object
response = requests.get('https://www.python.org')

# reponse.content possesses web page's composition
response.content 

# BeautifulSoup object is made to analyze text of response.content
soup = BeautifulSoup(response.content,'html5lib')

# text of website is returned without web browser structural info
text = soup.get_text(strip=True) 

# stops is equal to the stop words of the english language
stops = stopwords.words('english')

# textblob of'https://www.python.org'
blob = TextBlob(text)

# list to hold list of words that are not stop wrods
NonStopWords = []

# stop words are elminated from blob.words
# for loop appends words that are not stop words into the NonStopWords list
for wordFound in [word for word in blob.words if word not in stops]:
    NonStopWords.append(str(wordFound))
    

# wordcloud is generated from blob.words with no stop words
# .replace("'", '') is added to remove unecessary apostrophes
wordcloud = WordCloud(colormap='prism', background_color='white').generate(str(NonStopWords).replace("'", ''))

# plot axes are removed
plt.axis('off')

# input wordcloud to be shown
plt.imshow(wordcloud)

plt.show()


# Exercise 12.2

print('Exercise 12.2')
print()


print('Tokenized sentences of https://www.python.org')
print()

# print tokenized sentences of blob
print(blob.sentences)
print()

print('Tokenized words of https://www.python.org')
print()

# print tokenized words of blob
print(blob.words)
print()

print('Extracted noun phrases of https://www.python.org')
print()

# print extracted noun phrases of blob
print(blob.noun_phrases)
```
