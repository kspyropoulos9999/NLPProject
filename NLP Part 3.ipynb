```python
"""CIS 2532 NET 01
Dr. Shamsuddin
Kostas Spyropoulos
HW 7 
Program performs various uses of NLP
Exercise 12.4"""
import requests
import nltk
from nltk.corpus import stopwords
import imageio
from wordcloud import WordCloud
from textblob import TextBlob
import matplotlib.pyplot as plt
from textblob.sentiments import NaiveBayesAnalyzer
from bs4 import BeautifulSoup
from unidecode import unidecode
from textblob import Word
from spellchecker import SpellChecker

# Exercise 12.4
print()
print('Exercise 12.4')
print()

# the get function gets the URL argument and sends the URL's webpage as a Response object
response = requests.get('https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html')

# reponse.content possesses web page's composition
response.content

# BeautifulSoup object is made to analyze text of response.content
soup = BeautifulSoup(response.content,'html5lib')

# text of website is returned without web browser structural info
text = soup.get_text(strip=True)

# blob is equal to textblob of 'https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html'
# and will be analuzed with NaiveBayesAnalyzer
blobNaive = TextBlob(text, analyzer=NaiveBayesAnalyzer())

# print overall sentiment of blobNaive
print("Overall Naive Bayes Sentiment of https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html")
print(blobNaive.sentiment)
print()

print("Sentence Naive Bayes Sentiments of https://www.cnn.com/2022/03/25/entertainment/taylor-hawkins-foo-fighters-obit/index.html")

# print sentiment of every sentence in blobNaive
counter = 0
for sentenceNaive in blobNaive.sentences:
    counter = counter + 1
    # counter represents cardinal number of sentence sentiment
    print(str(counter) + str(". ")+ str(sentenceNaive.sentiment))
```
