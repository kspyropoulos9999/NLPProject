```python
"""CIS 2532 NET 01
Dr. Shamsuddin
Kostas Spyropoulos
HW 7 
Program performs various uses of NLP
Exercise 12.5"""
import requests
import nltk
from nltk.corpus import stopwords
import imageio
from wordcloud import WordCloud
from textblob import TextBlob
import matplotlib.pyplot as plt
from textblob.sentiments import NaiveBayesAnalyzer
from bs4 import BeautifulSoup
from unidecode import unidecode
from textblob import Word
from spellchecker import SpellChecker

# Exercise 12.5


print()
print('Exercise 12.5')
print()

# read the the Gutenberg.txt
with open('C:\\Users\\kspyr\\Documents\\Desktop Stuff\\CIS 2532\\HW 7\\Gutenberg.txt', 'r', encoding="utf8") as text1:
    content = text1.read();
    
# blob is equal to the textblob of the Gutenberg.txt
blob = TextBlob(content)

print("Misspelled words along with correction suggestions")
print()
# for loop iterate through all of the tokenized words of the blob and displays possible corrections for each misspelled word
for wordFound in blob.words:
    stringSpellCheck = str(Word(wordFound).spellcheck())
    stringZero = '0.0)'
    
    # if the spellcheck of the word of the blob being evaluated does not possess str(wordFound) and stringZero,
    # the word of the blob being evaluated must print the word of the blob as well as the correction suggestions.
    
    # str(wordFound) must be absent from the word of the blob being evaluated in order to execute the code under the if statement
    # because the presence of str(wordFound) in stringSpellCheck means the correction suggestion is the word itself which means the correction suggestion
    # is not truly a correction suggestion. In order to prevent such errors, str(wordFound) must be absent from the word of the blob being evaluated
    # to then execute the code under the if statement.
    
    # stringTwo must be absent from the word of the blob being evaluated in order to execute the code under the if statement
    # because the presence of stringTwo or '0.0)' in string represents that the correction suggestion's chance of being right is 0 which means
    # the supposed correction suggestion is not really a correction suggestion. In order to prevent such errors, '0.0)' must be absent from the word of the blob being evaluated
    # to then execute the code under the if statement.

    if str(wordFound) not in stringSpellCheck and stringZero not in stringSpellCheck:
        print()
        print('Word in error: ' + str(wordFound))
        print('Correction suggestions:')
        print(Word(wordFound).spellcheck())  
```
